```python
# gereken paketler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# veriyi okutalım 
import pandas as pd

df = pd.read_excel(r"C:\Users\arsla\python_project\customer_churn\WA_Fn-UseC_-Telco-Customer-Churn.xlsx")
```

```python
# ilk 5 satıra bakalım ve biraz bilgi alalım 

df.head()
df.describe()
```

```python
# yukarda baktığımızda sayısal veri görüyoruz bunu numeric olarak değiştiriyoruz

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
#  boş veri varmı diye kontrol ettik
df.isnull().sum()
```

```python
# boi veriyi temizledik

df = df.dropna()
```

```python
# charn sayılarını alalım
churn_counts = df["Churn"].value_counts()
# print(churn_counts)

# pie chart  yaplaım charns ları görmek için
plt.figure(figsize=(6,6))
plt.pie(churn_counts,
        labels = churn_counts.index,
        autopct='%1.1f%%', # Yüzdelik dilimleri gösterir (Örn: %26.5)
        startangle=140,    # Grafiği biraz döndürür, daha şık durur
        colors=['#66b3ff','#ff9999']) # Renkleri belirler (Mavi ve Kırmızımsı)
plt.title("churn dağıllımı")
plt.show()
```

```python
# şimdi biraz yorumlayalım 
#  müşterilerin kaçı bizi terk etmiş?
# yüksek faturanın etkisi
# boxplot  la buna biraz bakalım(iki değişkenin birbirine ilişkisi için en iyi seçenek)

plt.figure(figsize=(8,5))
sns.boxplot(x="Churn",y="MonthlyCharges",data=df)
plt.title("Aylık ödemelerin Churn durumuna göre dağılımı")
plt.show()
```

```python
# birazda cinsiyete bakalım burda frekanslara bakıp yorum yapabiliriz
sns.countplot(x="gender", hue="Churn",data = df ,palette ="pastel")
plt.title("Cinsiyet etkisi")
plt.show()
```
